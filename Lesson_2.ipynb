#### Задание 1

$ \overline{a} = 
\begin{pmatrix}
     20     \\
     20     \\
     20  \\
\end{pmatrix}
\quad
\overline{b} = 
\begin{pmatrix}
     0     \\
     0     \\
     -20  \\
\end{pmatrix}$

$\overline{a} + \overline{b} = 
\begin{pmatrix}
     20     \\
     20    \\
     0  \\
\end{pmatrix}$

```python
import math

def vect_len(a):
    sum_2 = 0
    for i in a:
        sum_2 += i**2
    return math.sqrt(sum_2)

a = [3,4]

print(f'Длина вектора равна {vect_len(a):.3f}')
```
