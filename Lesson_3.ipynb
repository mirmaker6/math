#### Задание 1

Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
k = [0.5, 1, 3]
a = [1, 3, 4]
b = [3, 2, 1]
color = ['red', 'green', 'gray']

for i in range(len(k)):
    
    y = k[i] * (np.cos(x - a[i])) + b[i]
    plt.plot(x ,y, color=color[i])

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

#### Задание 4

1)Решите систему уравнений:

y = x^2 – 1

exp(x) + x∙(1 – y) = 1

2)Решите систему уравнений и неравенств:

y = x^2 – 1

exp(x) + x∙(1 – y) > 1


1)

```python
x = np.linspace(-5, 5, 201)
plt.plot(x , (np.exp(x) + x - 1)/x)
plt.plot(x , x**2 - 1)

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
from scipy.optimize import fsolve

def equation(p):
    x ,y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1 , y1 = fsolve(equation, (-2, 4))
x2 , y2 = fsolve(equation, (2, 5))
x3 , y3 = fsolve(equation, (4, 15))
print(f'point 1\t x1={x1:.2f} \ty1={y1:.2f}\n\
point 2\t x2={x2:.2f} \ty2={y2:.2f}\n\
point 3\t x3={x3:.2f} \ty3={y3:.2f}')
```

2)

```python
x_1 = np.linspace(-5, x1, 201)
x_2 = np.linspace(x2, x3, 201)

plt.plot(x_1, x_1**2 - 1)

plt.plot(x_2, x_2**2 - 1)

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
